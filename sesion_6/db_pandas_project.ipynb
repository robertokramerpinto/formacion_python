## Counts and GroupBY Operations

```python
titanic_df["Sex"].value_counts()
```

```python
titanic_df["Sex"].value_counts(normalize=True)
```

```python
titanic_df["Cabin"].value_counts(normalize=True,dropna=False)
```

```python
# Create un group object
grouped = titanic_df.groupby(["Sex"])
grouped
```

```python
grouped.mean()
```

```python
grouped[["Survived","Age"]].mean()
```

```python
titanic_df.groupby(["Sex"])[["Survived","Age"]].mean()
```

```python
agg_df = titanic_df.groupby(["Sex"])[["Survived","Age"]].mean()
agg_df
```

```python
agg_df.index
```

```python
titanic_df.groupby(["Sex"],as_index=False)[["Survived","Age"]].mean()
```

### multiple agg functions

```python
functions = ["mean","max","count"]
titanic_df.groupby(["Sex"])[["Survived","Age"]].agg(functions)
```

```python
# use different functions across different columns
titanic_df.groupby(["Sex"]).agg({
    "Survived":["mean"],
    "Age":["count","mean"],
    "Fare":["mean","max","min"]
})
```
